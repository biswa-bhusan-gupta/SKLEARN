```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score,f1_score
```

```python
def Evaluate_Pred(Y_Actual,Y_Pred):
    Accuracy=accuracy_score(Y_Actual,Y_Pred)
    Precision_Score=precision_score(Y_Actual,Y_Pred)
    Recall_Score=recall_score(Y_Actual,Y_Pred)
    F1_Score=f1_score(Y_Actual,Y_Pred)
    print(f'Confusion Matrix:\n{confusion_matrix(Y_Actual,Y_Pred)}')
    print(f'Accuracy Score:{Accuracy*100:.2f}%')
    print(f'Classificatio Report:\n{classification_report(Y_Actual,Y_Pred)}')
    print(f'Presion Score:{Precision_Score:.2f}')
    print(f'Recall Score:{Recall_Score:.2f}')
    print(f'F1 Score:{F1_Score:.2f}')
```

```python
RFC.fit(X_Train,Y_Train)
Y_Pred=RFC.predict(X_Valid)
```

```python
Y_Pred
```

```python
Evaluation1=Evaluate_Pred(Y_Valid,Y_Pred)
```

```python
RFC1=RandomForestClassifier(n_estimators=85)
RFC1.fit(X_Train,Y_Train)
Y_Pred1=RFC1.predict(X_Valid)
```

```python
Y_Pred1
```

```python
Evaluation2=Evaluate_Pred(Y_Valid,Y_Pred1)
```

```python
# RANDOMIZED SEARCH CV:
from sklearn.model_selection import RandomizedSearchCV
```

```python
Grid={'n_estimators':[10,100,200,500,1000,1200],
      'max_depth':[None,5,10,20,30],
      'max_features':['auto','sqrt'],
      'min_samples_split':[2,4,6],
      'min_samples_leaf':[1,2,4]}
```

```python
(6*5*2*3*3)*5
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
RFC2=RandomForestClassifier(n_jobs=1)
RSC=RandomizedSearchCV(estimator=RFC2,param_distributions=Grid,n_iter=10,cv=5,verbose=2)
# ALL ARE SELECTED RANDOMLY(CV=5,N-ITER=10--->5*10=50FITS) (10 RANDOMLY CHOOSEN USING 5 PARAMETERS IN GRID)
```

```python
RSC.fit(x_train,y_train)
```

```python
RSC.best_params_
```

```python
y_pred=RSC.predict(x_test)
```

```python
y_pred
```

```python
Evaluation3=Evaluate_Pred(y_test,y_pred)
```

```python
# GRID SEARCH CV:
Grid1={'n_estimators':[100,200,500],
      'max_depth':[None],
      'max_features':['auto','sqrt'],
      'min_samples_split':[6],
      'min_samples_leaf':[1,2]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
RFC3=RandomForestClassifier(n_jobs=1)
GSC=GridSearchCV(estimator=RFC3,param_grid=Grid1,cv=5,verbose=2)
```

```python
GSC.fit(x_train,y_train)
```

```python
GSC.best_params_
```

```python
y_pred1=GSC.predict(x_test)
y_pred1
```

```python
Evaluation4=Evaluate_Pred(y_test,y_pred1)
```
